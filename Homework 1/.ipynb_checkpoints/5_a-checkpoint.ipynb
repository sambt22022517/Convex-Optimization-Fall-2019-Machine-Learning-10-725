## 5 Solving optimization problems with CVX

### Chuẩn bị

```python
import cvxpy as cp
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
xy_train = pd.read_csv('xy_train.csv', 
                       header=None, 
                       names=['Feature 1', 'Feature 2', 'Label'])
xy_test = pd.read_csv('xy_test.csv', 
                       header=None, 
                       names=['Feature 1', 'Feature 2', 'Label'])
```

```python
X = np.array(xy_train[['Feature 1', 'Feature 2']])
y = np.array(xy_train['Label'])

X_test = np.array(xy_test[['Feature 1', 'Feature 2']])
y_test = np.array(xy_test['Label'])

n, p = X.shape
```

### Phần a 



### Phần a.i



```python
beta = cp.Variable(p)
beta0 = cp.Variable(1)
epsilon = cp.Variable(n)
obj = cp.Minimize(1/2 * cp.norm(beta) ** 2 + cp.sum(epsilon))
constraints = [epsilon >= 0,
              cp.multiply(y, (X @ beta + beta0)) >= 1 - epsilon]
prob = cp.Problem(obj, constraints)
print("Optimal value", prob.solve(solver=cp.ECOS))
print("Beta", beta.value)
print("Beta0", beta0.value)
```

### Phần a.ii



```python
x0 = min(X[:,0])
x1 = max(X[:,0])
y0 = (- beta0.value[0] - beta.value[0] * x0) / beta.value[1]
y1 = (- beta0.value[0] - beta.value[0] * x1) / beta.value[1]

sns.scatterplot(x=xy_train['Feature 1'], y=xy_train['Feature 2'], hue=xy_train['Label'])
plt.plot([x0, x1], [y0, y1])
```

### Phần a.iii



```python
XCopy = X.copy()
XCopy[:, 0] = cp.multiply(y, X[:, 0]).value
XCopy[:, 1] = cp.multiply(y, X[:, 1]).value
w = cp.Variable(n)
obj = cp.Maximize(-1/2 * cp.sum_squares(XCopy.T @ w) + np.ones(n).T @ w)
constraints = [0 <= w, w <= 1,
               w.T @ y == 0]
prob = cp.Problem(obj, constraints)
print(prob.solve(solver=cp.ECOS))
```

### Phần a.iv



```python
beta = cp.Variable(p)
beta0 = cp.Variable(1)
epsilon = cp.Variable(n)
a = cp.Variable(1)
```

```python
result = [list(range(-5, 6)), []]
for i in range(-5, 6):
    C = 2 ** i
    obj = cp.Minimize(1/2 * cp.norm(beta) ** 2 + C * cp.sum(epsilon))
    constraints = [epsilon >= 0,
                  cp.multiply(y, (X @ beta + beta0)) >= 1 - epsilon]
    prob = cp.Problem(obj, constraints)
    prob.solve(solver=cp.ECOS)
    
    y_i = beta0.value + X_test @ np.array(beta.value).T >= 0
    y_i = y_i * 2 - 1
    result[1].append(y_test.shape[0]-sum(y_i == y_test))
```

```python
sns.scatterplot(x=result[0], y=result[1])
result[0], result[1]
```
